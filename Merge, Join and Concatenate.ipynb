**Popis typů spojení:**

- `inner` – pouze společné řádky z obou tabulek
- `left` – všechny řádky z levé tabulky, pokud je shoda v pravé, připojí se
- `right` – opačně než left
- `outer` – spojí všechno z obou tabulek, chybějící doplní jako NaN

```python
df1.merge(df2)  #  #  Vnitřní spojení (inner join) - pouze shodné záznamy
```

```python
df1.merge(df2, how='inner', on='flavor_id')  #  Vnitřní spojení (inner join) s určením sloupce pro spojení
```

```python
df1.merge(df2, how='inner', on=['flavor_id', 'flavor_name'])  #  Vnitřní spojení s více sloupci
```

```python
df1.merge(df2, how='outer')  #  Vnější spojení (outer join) - všechny záznamy z obou tabulek
```

```python
df1.merge(df2, how='left')  #  Levé spojení (left join) - všechny záznamy z levé tabulky a odpovídající z pravé
```

```python
df1.merge(df2, how='right')  #  Pravé spojení (right join) - všechny záznamy z pravé tabulky a odpovídající z levé
```

```python
df1.merge(df2, how='cross')  #  Křížové spojení (cross join) - všechny kombinace záznamů z obou tabulek
```

```python
df3=df1.set_index('flavor_id').join(df2.set_index('flavor_id'), lsuffix='_L', rsuffix='_R', how= 'outer') #  Spojení pomocí indexu s přidáním suffixů a vnějšího spojení
df3
```

```python
import pandas as pd  
pd.concat([df1, df2])  #  Spojení dvou datových rámců pod sebe (vertikální spojení)
```

```python
pd.concat([df1, df2],join='inner')  #  Spojení dvou datových rámců pod sebe s vnitřním spojením (pouze shodné sloupce)
```

```python
pd.concat([df1, df2],join='outer')  #  Spojení dvou datových rámců pod sebe s vnějším spojením (všechny sloupce)
```

```python
pd.concat([df1, df2],join='outer', axis=1)  #  Spojení dvou datových rámců vedle sebe (horizontální spojení)
```

```python
df1.append(df2)  #  Přidání datového rámce df2 na konec datového rámce df1 (vertikální spojení)
```

.append() byl považován za zastaralý a pomalý – pd.concat() je rychlejší a univerzálnější.
