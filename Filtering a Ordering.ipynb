```python
df2.filter(items= ['Country','Population','Area (sq. mi.)'], axis=1)  # vybereme jenom sloupce Country, Population a Area (sq. mi.)
```

</p>


```python
df2.filter(items= ['Country','Population','Area (sq. mi.)'], axis=0) # vybereme jenom radky Country, Population a Area (sq. mi.)
```

<p class="cs-info">
  Nejprve převedeme hodnoty ve sloupci <code>GDP ($ per capita)</code> na číselný typ pomocí <code>pd.to_numeric()</code>. 
  Pokud se v datech nachází nečíselné hodnoty (např. „–“), budou automaticky nahrazeny <code>NaN</code>.
</p>

<p class="cs-info">
  Následně vytvoříme kopii původního DataFrame jako <code>df_rank</code>, abychom zachovali původní data beze změn.
</p>

<p class="cs-info">
  Do kopie přidáme nový sloupec <code>Rank</code>, který přiřadí pořadí zemím podle výše HDP na obyvatele (sestupně).
  Nejvyšší HDP získá hodnocení 1. Výsledky jsou zaokrouhleny a typ sloupce je <code>Int64</code>, což umožňuje i hodnoty <code>NaN</code>.
</p>

<p class="cs-info">
  Nakonec přeuspořádáme sloupce tak, aby se <code>Rank</code> zobrazoval jako první, a výsledný DataFrame byl přehlednější.
</p>

```python
# Převeďme GDP na čísla v původním DataFrame
df['GDP ($ per capita)'] = pd.to_numeric(df['GDP ($ per capita)'], errors='coerce')
# Vytvořme nový DataFrame, aby původní zůstal nezměněn (volitelné)
df_rank = df.copy()
# Vytvoř Rank
df_rank['Rank'] = df_rank['GDP ($ per capita)'].rank(ascending=False).round().astype('Int64')
# Přesuň Rank na první místo
df_rank = df_rank[['Rank'] + [col for col in df_rank.columns if col != 'Rank']]
# Výstup
df_rank
```

```python
df_rank.to_csv('World population.csv', index=False)  # ulozime do csv souboru
```

```python
df_rank[df_rank['Rank'] <= 10].sort_values(by='Rank', ascending=True)  # vybereme top 10 zemí podle HDP na hlavu a seřadíme je podle Rank
```

```python
df_rank[df_rank['Rank'] <= 10].sort_values(by=['Rank','Country'], ascending=[False, True])  # vybereme top 10 zemí podle HDP na hlavu a seřadíme je podle Rank sestupně a podle Country vzestupně
```
