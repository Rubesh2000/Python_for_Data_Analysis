```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Vytvoř 10 dat (posledních 10 dnů)
dates = [datetime.today() - timedelta(days=i) for i in range(9, -1, -1)]
dates = [d.strftime('%Y-%m-%d') for d in dates]

# Vygeneruj náhodná čísla < 1 (a zaokrouhli na 5 desetinných míst)
def random_column():
    return np.round(np.random.uniform(0.00001, 0.99999, 10), 5)

# Vytvoření DataFrame
df = pd.DataFrame({
    'Date': dates,
    'Flavor Rating': random_column(),
    'Texture Rating': random_column(),
    'Overall Rating': random_column()
})

# Výpis
print(df)

# Uložení do CSV
df.to_csv("Ice Cream Ratings.csv", index=False, float_format='%.5f')

```

<p class="cs-info">
  <code>df.plot(kind='line', subplots=True)</code> vytvoří samostatný čárový graf pro každý číselný sloupec v DataFrame <code>df</code>.

```python
df.plot(kind= 'line', subplots=True) # subplots=True vytvoří samostatné grafy pro každý sloupec
```
