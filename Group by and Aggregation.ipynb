Group by and Aggregation

```python
import pandas as pd
import numpy as np
import random

# Základní možnosti
flavors = [
    "Mint Chocolate Chip", "Chocolate", "Vanilla", "Cookie Dough", "Rocky Road",
    "Pistachio", "Cake Batter", "Neapolitan", "Chocolate Fudge Brownie", "Strawberry",
    "Caramel Swirl", "Mango", "Coffee", "Blueberry", "Butter Pecan"
]
base_flavors = ["Vanilla", "Chocolate"]
liked_options = ["Yes", "No"]

# Generování dat
data = []
for _ in range(50):
    flavor = random.choice(flavors)
    base = random.choice(base_flavors)
    liked = random.choice(liked_options)
    flavor_rating = round(np.random.uniform(2.0, 10.0), 1)
    texture_rating = round(np.random.uniform(2.0, 10.0), 1)
    total_rating = round(flavor_rating + texture_rating, 1)
    data.append([flavor, base, liked, flavor_rating, texture_rating, total_rating])

# Vytvoření DataFrame
df = pd.DataFrame(data, columns=[
    "Flavor", "Base Flavor", "Liked", "Flavor Rating", "Texture Rating", "Total Rating"
])

# Uložení do CSV
df.to_csv("Flavor.csv", index=False)
```

```python
df=pd.read_csv(r"data\Flavor.csv")
df.head()
```

```python
df. info()  # Zobrazí informace o DataFrame
```

```python
df.groupby('Base Flavor') #ukaže jaké jsou hodnoty v Base Flavor (text)
```

```python
grouped= df.groupby('Base Flavor') # pridam grouped, abych mohl použít další funkce
```

```python
grouped.mean(numeric_only=True)  # Vypočítá průměrné hodnoty pro každou skupinu ....To řekne Pandas, ať ignoruje nečíselné sloupce
```

```python
df.groupby('Base Flavor').count()    #count() spočítá počet řádků v každé skupině (text a čísla)
```

```python
df.groupby('Base Flavor').min()  # Vypočítá minimální hodnoty pro každou skupinu   Pro Liked , A protože 'No' < 'Yes', vybere No (menší než 'Yes')jako minimální hodnotu.
```

```python
df.groupby('Base Flavor').max() # Vypočítá maximální hodnoty pro každou skupinu  Pro Liked , A protože 'Yes' > 'No', vybere Yes (větší než 'No')jako maximální hodnotu.
```

```python
df.groupby('Base Flavor').sum(numeric_only=True)  # Vypočítá součet hodnot pro každou skupinu (pouze číselné sloupce)
```

```python
df.groupby('Base Flavor').agg({ 'Flavor Rating': 'mean', 
                               'Texture Rating': 'sum', 
                               'Total Rating': 'max'}) # Vypočítá průměrnou hodnotu 'Flavor Rating', součet 'Texture Rating' a maximální hodnotu 'Total Rating' pro každou skupinu
```

```python
df.groupby('Base Flavor').agg({ 'Flavor Rating': ['mean', 'min', 'max', 'count','sum']})  #  Vypočítá průměr, minimum, maximum, počet a součet 'Flavor Rating' pro každou skupinu
```

```python
df.groupby('Base Flavor').agg({ 'Flavor Rating': ['mean', 'min', 'max', 'count','sum'],
                               'Texture Rating': ['mean', 'min', 'max', 'count','sum'],
                               'Total Rating':['mean', 'min', 'max', 'count','sum']})  
```

```python
df.groupby(['Base Flavor', 'Liked']). mean()   #Ne funguje, protože 'Base Flavor' a 'Liked' je textový sloupec, ale funguje s numerickými sloupci
```

```python
df.groupby(['Base Flavor', 'Liked']).agg({ 'Flavor Rating': ['mean', 'min', 'max', 'count','sum']})
```
