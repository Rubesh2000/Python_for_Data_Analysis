<p class="cs-info">

<p class="cs-info">

</p>


```python
import pandas as pd
df = pd.read_csv(r"data\World population.csv", dtype={'Rank': 'Int64'})  # Nacti CSV soubor do DataFrame, specifikuj typ pro sloupec 'Rank'
df
```

```python
print([i for i in df.index if i != i.strip()]) # Kontrola !!! Vypíše seznam indexů, které mají mezery na začátku nebo konci
```

```python
df.index = df.index.str.strip()  # Odstraní mezery na začátku a konci indexu
print([i for i in df.index if i != i.strip()])  # mělo by vrátit prázdný seznam
```

<p class="cs-info">
  Tento příkaz upravuje <strong>víceúrovňový index (MultiIndex)</strong> v objektu <code>df</code>.
  Pomocí <code>set_levels()</code> odstraní mezery na začátku a konci každé hodnoty v jednotlivých úrovních indexu.
</p>

<p class="cs-info">
  Konkrétně volá metodu <code>str.strip()</code> pro každou úroveň indexu v <code>df.index.levels</code>, čímž se zajistí, že všechny hodnoty indexu budou čisté, bez nechtěných mezer.
</p>

<p class="cs-info">
  Výsledkem je konzistentní a přesnější práce s daty, protože hodnoty v indexu nebudou obsahovat bílé znaky, které by mohly způsobit chyby při filtrování, spojování nebo seskupování dat.
</p>


```python
df.index = df.index.set_levels([level.str.strip() for level in df.index.levels]) # Odstraní mezery na začátku a konci indexu v MultiIndex
```

```python
# Ověříme, jestli tam je požadovaný záznam
print([i for i in df.index if 'Norway' in i])
```
